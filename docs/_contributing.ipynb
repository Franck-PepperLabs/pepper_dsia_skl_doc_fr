# [**Contribuer**](https://scikit-learn.org/stable/developers/contributing.html)

Ce projet est une initiative communautaire et tout le monde est invité à contribuer.

Le projet est hébergé sur https://github.com/scikit-learn/scikit-learn

Le processus de prise de décision et la structure de gouvernance de scikit-learn sont exposés dans le document de gouvernance : [Gouvernance et prise de décision de scikit-learn](https://scikit-learn.org/stable/governance.html#governance).

Scikit-learn est quelque peu [sélectif](https://scikit-learn.org/stable/faq.html#selectiveness) lorsqu'il s'agit d'ajouter de nouveaux algorithmes, et la meilleure façon de contribuer et d'aider le projet est de commencer à travailler sur des problèmes connus. Consultez [Problèmes pour les nouveaux contributeurs](https://scikit-learn.org/stable/developers/contributing.html#new-contributors) pour commencer.

**Notre communauté, nos valeurs**

Nous sommes une communauté basée sur l'ouverture et des discussions amicales et didactiques.

Nous aspirons à traiter tout le monde de manière équitable et à valoriser leurs contributions. Nous recherchons particulièrement des personnes issues de milieux sous-représentés dans les logiciels Open Source et en particulier dans scikit-learn, afin de participer et de contribuer avec leur expertise et leur expérience.

Les décisions sont prises en fonction du mérite technique et du consensus.

Le code n'est pas la seule façon d'aider le projet. Examiner les demandes de fusion, répondre aux questions pour aider les autres sur les listes de diffusion ou les problèmes, organiser et enseigner des tutoriels, travailler sur le site web, améliorer la documentation, sont toutes des contributions inestimables.

Nous adhérons aux principes d'ouverture, de respect et de considération des autres de la Python Software Foundation : https://www.python.org/psf/codeofconduct/

Si vous rencontrez des problèmes lors de l'utilisation de ce package, n'hésitez pas à soumettre un ticket sur le [suivi des problèmes de GitHub](https://github.com/scikit-learn/scikit-learn/issues). Vous êtes également invité à soumettre des demandes de fonctionnalités ou des demandes de fusion.

# Moyens de contribuer

Il existe de nombreuses façons de contribuer à scikit-learn, les plus courantes étant la contribution de code ou de documentation au projet. Améliorer la documentation n'est pas moins important que d'améliorer la bibliothèque elle-même. Si vous trouvez une faute de frappe dans la documentation ou si vous avez apporté des améliorations, n'hésitez pas à envoyer un e-mail à la liste de diffusion ou de préférence à soumettre une demande de fusion sur GitHub. La documentation complète se trouve dans le répertoire doc/.

Mais il existe de nombreuses autres façons d'aider. En particulier, aider à [améliorer, trier et enquêter sur les problèmes](https://scikit-learn.org/stable/developers/bug_triaging.html#bug-triaging) et [examiner les demandes de fusion d'autres développeurs](https://scikit-learn.org/stable/developers/contributing.html#code-review) sont des contributions très précieuses qui allègent la charge des mainteneurs du projet.

Une autre façon de contribuer consiste à signaler les problèmes que vous rencontrez et à approuver les problèmes signalés par d'autres qui vous concernent. Cela nous aide également si vous en parlez : référencez le projet sur votre blog et dans des articles, faites un lien vers celui-ci à partir de votre site web ou simplement étoilez pour dire "je l'utilise" :

---

Dans le cas où une contribution/un problème implique des modifications des principes de l'API ou des modifications des dépendances ou des versions prises en charge, elle doit être appuyée par une [proposition d'amélioration (SLEP)](https://scikit-learn.org/stable/governance.html#slep), où un SLEP doit être soumis sous forme de demande de fusion à [propositions d'amélioration](https://scikit-learn-enhancement-proposals.readthedocs.io/) en utilisant le [modèle SLEP](https://scikit-learn-enhancement-proposals.readthedocs.io/en/latest/slep_template.html) et suivre le processus de prise de décision décrit dans [Gouvernance et prise de décision de scikit-learn](https://scikit-learn.org/stable/governance.html#governance).

# Soumettre un rapport de bogue ou une demande de fonctionnalité

Nous utilisons les problèmes de GitHub pour suivre tous les bugs et demandes de fonctionnalités ; n'hésitez pas à ouvrir un problème si vous avez trouvé un bogue ou souhaitez voir une fonctionnalité implémentée.

Dans le cas où vous rencontrez des problèmes lors de l'utilisation de ce package, n'hésitez pas à soumettre un ticket au [Bug Tracker](https://github.com/scikit-learn/scikit-learn/issues). Vous êtes également invité à soumettre des demandes de fonctionnalités ou des demandes de fusion.

Il est recommandé de vérifier que votre problème est conforme aux règles suivantes avant de le soumettre :
- Vérifiez que votre problème n'est pas déjà traité par d'autres [problèmes](https://github.com/scikit-learn/scikit-learn/issues?q=) ou [demandes de fusion](https://github.com/scikit-learn/scikit-learn/pulls?q=).
- Si vous soumettez une demande d'algorithme ou de fonctionnalité, vérifiez que l'algorithme répond à nos [critères d'inclusion des nouveaux algorithmes](https://scikit-learn.org/stable/faq.html#what-are-the-inclusion-criteria-for-new-algorithms).
- Si vous soumettez un rapport de bogue, nous vous encourageons vivement à suivre les directives de [Comment faire un bon rapport de bogue](https://scikit-learn.org/stable/developers/contributing.html#filing-bugs).

## Comment faire un bon rapport de bogue

Lorsque vous soumettez un problème sur [Github](https://github.com/scikit-learn/scikit-learn/issues), veuillez faire de votre mieux pour suivre ces directives ! Cela facilitera grandement la fourniture de commentaires pertinents :

- Le rapport de bogue idéal contient un [court extrait de code reproductible](https://scikit-learn.org/stable/developers/minimal_reproducer.html#minimal-reproducer), de cette manière, n'importe qui peut essayer de reproduire facilement le bogue (voir [ici](https://stackoverflow.com/help/mcve) pour plus de détails). Si votre extrait de code est plus long d'environ 50 lignes, veuillez fournir un lien vers un [gist](https://gist.github.com/) ou un dépôt GitHub.
- Si cela n'est pas possible d'inclure un extrait de code reproductible, veuillez être précis sur les **estimateurs et/ou les fonctions impliqués et la forme des données**.
- Si une exception est levée, veuillez **fournir la trace complète**.
- Veuillez inclure votre **type et numéro de version du système d'exploitation**, ainsi que vos **versions de Python, scikit-learn, numpy et scipy**. Ces informations peuvent être obtenues en exécutant le code suivant :
```python
import sklearn
sklearn.show_versions()
```
- Veuillez vous assurer que tous les **extraits de code et messages d'erreur sont formatés dans des blocs de code appropriés**. Consultez [Créer et mettre en évidence des blocs de code](https://help.github.com/articles/creating-and-highlighting-code-blocks) pour plus de détails.

Si vous souhaitez aider à la gestion des problèmes, lisez [ce qui suit](https://scikit-learn.org/stable/developers/bug_triaging.html#bug-triaging).
